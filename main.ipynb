## Informing about severity

```python
# Imports
import pandas as pd


# Load codebook data
codebook = pd.read_csv('./data/codebook.csv')

# Define relevant visualization types which we are interested in
relevant_map_types = ["Choropleth Map", "Bubble Map"]
relevant_bar_chart_types = ["Classic Barchart", "Simple Stacked Barchart", "Diverging Stacked Barchart", "Spanchart"]
relevant_line_chart_types = ["Classic Linechart", "Areachart", "Stacked Areachart", "Streamgraph"]
relevant_other_types = ["Heatmap", "Piechart", "Treemap", "Scatterplot", "Network", "Bubblechart", "Flowchart", "Radar"]
relevant_visualization_types = [*relevant_map_types, *relevant_bar_chart_types, *relevant_line_chart_types, *relevant_other_types]

# Define intended message with subcodes
informing_about_severity = ["Magnitude Spread", "Supplies, Preparedness Capacities", "Historical comparison"]

# Create a new data frame with only the colums for the relevant visualizations
relevant_columns = [*relevant_visualization_types, *informing_about_severity]
reduced_codebook = codebook[relevant_columns]

# Only keep rows for which at least one column for the intended message (e.g informing about severity) is True
keep_row_filter = reduced_codebook[informing_about_severity].any(axis="columns").to_numpy().astype(bool)
reduced_codebook = reduced_codebook.iloc[lambda x: keep_row_filter[x.index] == True]

# Create new data frame which only holds the counts for the relevant visualization types in a single row
visualization_types_count = []

for column in relevant_visualization_types:
    counts = reduced_codebook[column].value_counts()
    true_counts = counts.get(True, 0)
    visualization_types_count.append(true_counts)

df = pd.DataFrame(
    {
        "visualization_type": relevant_visualization_types,
        "count": visualization_types_count
    },
    index=relevant_visualization_types
    )
df = df.sort_values('count')
df.plot(kind="bar", title="Informing about severity", ylabel="Anzahl der Verwendungen", xlabel = "", x="visualization_type", y="count", colormap="Paired", legend = False)
```

## Communicating risk, vulnerability and equity

```python
# Imports
import pandas as pd


# Load codebook data
codebook = pd.read_csv('./data/codebook.csv')

# Define relevant visualization types which we are interested in
relevant_map_types = ["Choropleth Map", "Bubble Map"]
relevant_bar_chart_types = ["Classic Barchart", "Simple Stacked Barchart", "Diverging Stacked Barchart", "Spanchart"]
relevant_line_chart_types = ["Classic Linechart", "Areachart", "Stacked Areachart", "Streamgraph"]
relevant_other_types = ["Heatmap", "Piechart", "Treemap", "Scatterplot", "Network", "Bubblechart", "Flowchart", "Radar"]
relevant_visualization_types = [*relevant_map_types, *relevant_bar_chart_types, *relevant_line_chart_types, *relevant_other_types]

# Define intended message with subcodes
communicating_risk_vulnerability_equity = ["Health Equity & Vulnerability"]

# Create a new data frame with only the colums for the relevant visualizations
relevant_columns = [*relevant_visualization_types, *communicating_risk_vulnerability_equity]
reduced_codebook = codebook[relevant_columns]

# Only keep rows for which at least one column for the intended message (e.g informing about severity) is True
keep_row_filter = reduced_codebook[communicating_risk_vulnerability_equity].any(axis="columns").to_numpy().astype(bool)
reduced_codebook = reduced_codebook.iloc[lambda x: keep_row_filter[x.index] == True]

# Create new data frame which only holds the counts for the relevant visualization types in a single row
visualization_types_count = []

for column in relevant_visualization_types:
    counts = reduced_codebook[column].value_counts()
    true_counts = counts.get(True, 0)
    visualization_types_count.append(true_counts)

df = pd.DataFrame(
    {
        "visualization_type": relevant_visualization_types,
        "count": visualization_types_count
    },
    index=relevant_visualization_types
    )
df = df.sort_values('count')
df.plot(kind="bar", title="Communicating risk, vulnerability and equity", ylabel="Anzahl der Verwendungen", xlabel = "", x="visualization_type", y="count", colormap="Paired", legend = False)
```
